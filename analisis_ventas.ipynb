# 📊 Análisis de Ventas Online

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from src.utils import cargar_datos, agregar_columna_total

df = cargar_datos("../data/ventas.csv")
df = agregar_columna_total(df)
print(df.head())

# Ventas totales por categoría
ventas_categoria = df.groupby("categoria")["total"].sum().reset_index()
print(ventas_categoria)

# Ventas mensuales
df["mes"] = df["fecha"].dt.to_period("M")
ventas_mensuales = df.groupby("mes")["total"].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=ventas_mensuales, x="mes", y="total", palette="Blues")
plt.title("Ventas Totales por Mes")
plt.xlabel("Mes")
plt.ylabel("Ventas ($)")
plt.savefig("../reports/ventas_mensuales.png", dpi=300)
plt.show()
```
